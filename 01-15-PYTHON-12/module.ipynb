### Задание 1.2 ###

Преобразовать все столбцы, в которых меньше 150 уникальных значений, в тип данных `category`, исключив из преобразования столбцы **`Date`**, **`Rooms`**, **`Bedroom`**, **`Bathroom`**, **`Car`**.

В качестве ответа записать результирующее количество столбцов, которые имеют тип данных `category`.

***Ответ - 10***

```python
unique_list = []
for col in melb_df.columns:
    item = (col, melb_df[col].nunique(), melb_df[col].dtypes)
    unique_list.append(item)
unique_counts = pd.DataFrame(
    unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type']
).sort_values(by='Num_Unique', ignore_index=True)
display(unique_counts)
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_count = 150
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
display(melb_df.info())
```

###  Задание 3.3 ###

Какая риелторская компания (`SellerG`) имеет наименьшую общую выручку за период с 1 мая по 1 сентября (включительно) 2017 года?

Для ответа на этот вопрос рассчитать сумму продаж (`Price`) каждой компании в заданный период.

# Сводные таблицы #

## Метод groupby как способ построения сводных таблиц ##

```python
melb_df.groupby('Rooms')[['Price', 'BuildingArea']].median()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack()
```

## Метод pivot_table для построения сводных таблиц ##

```python
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
```

```python
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count'
)
```

```python
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

## Многомерные сводные таблицы ##

```python
melb_df.pivot_table(
    values='Price',
    index=['Method', 'Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)
```

## Доступ к данным в сводной таблице ##

```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
pivot.columns
```

```python
display(pivot)
display(pivot['mean'])
display(pivot['mean']['unit'])
```

```python
mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
display(filtered_pivot)

print(list(filtered_pivot.index))
```
