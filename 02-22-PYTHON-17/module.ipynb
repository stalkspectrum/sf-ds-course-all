###  Задание 5.4 ###

Написать функцию `wiki_header`, которая по адресу страницы возвращает заголовок для статей на Wikipedia.

```python
def wiki_header(wikiurl):
    response = requests.get(wikiurl)
    page = BeautifulSoup(response.text, 'html.parser')
    return page.find('h1').text
```

```python
wiki_header('https://en.wikipedia.org/wiki/Operating_system')
```

### Сбор нескольких элементов: собираем все ссылки на странице ###

```python
url = 'https://en.wikipedia.org/wiki/List_of_programming_languages'
response = requests.get(url)
page = BeautifulSoup(response.text, 'html.parser')
print(page.find('a'))
```

```python
links = page.find_all('a')
print(len(links))
```

```python
print([link.text for link in links[500:510]])
```

```python
print([link.text for link in links[0:10]])
```

## Как настроить автоматический запуск ##

```python
import schedule
import time
```

```python
def task():
    print('Hello! I am a task.')
    return None
```

```python
schedule.every(10).seconds.do(task)
```

```python
n = 0
while n < 5:
    n += 1
    schedule.run_pending()
    time.sleep(5)
```
